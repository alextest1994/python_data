```python
import pandas as pd
from pyjarowinkler import distance
import Levenshtein as lev
```

```python
amazon = pd.read_csv("./amazon_tr.csv")
walmart = pd.read_csv("./walmart_tr.csv")
online_shop = pd.read_csv("./online_shop_tr.csv")
```

```python
def jaccard_similarity(list1, list2):
    intersection = []
    union = []
    duple = []
    intersection = len(list(set(list1).intersection(list2))) 
    print(intersection)
    duple = len((list1)) - len(set(list1)) + len((list2)) - len(set(list2)) 
    union = (len(list1) + len(list2)) - intersection 
    print(union)    
    print(len((list1)))
    print(set((list1)))
    print(len(set((list1))))
    print(len((list2)))
    print(set((list2)))
    print(len(set((list1))))
    print(duple)
    return float(intersection) / union
```

```python
online_shop = online_shop.head(200)
```

```python
result_frame = pd.MultiIndex.from_product(
    [online_shop["index"], amazon["index"]], names=["online_shop", "amazon"]).to_frame(index=False)
```

```python
result_frame
```

```python
result_frame["ama_jarowinkler_sim"] = [distance.get_jaro_distance(i,j) for i,j in zip(result_frame["online_shop"],result_frame["amazon"])]
result_frame["ama_levenshtein_sim"] = [lev.ratio(i,j) for i,j in zip(result_frame["online_shop"],result_frame["amazon"])]
result_frame["ama_jaccard_sim"] = [jaccard_similarity(i,j) for i,j in zip(result_frame["online_shop"],result_frame["amazon"])]
```

```python
result_frame.head(200)
```

```python

max_value_jaro_am = result_frame["ama_jarowinkler_sim"].max()
min_value_jaro_am = result_frame["ama_jarowinkler_sim"].min()
mean_value_jaro_am = result_frame["ama_jarowinkler_sim"].mean()
max_value_lev_am = result_frame["ama_levenshtein_sim"].max()
min_value_lev_am = result_frame["ama_levenshtein_sim"].min()
mean_value_lev_am = result_frame["ama_levenshtein_sim"].mean()
max_value_jac_am = result_frame["ama_jaccard_sim"].max()
min_value_jac_am = result_frame["ama_jaccard_sim"].min()
mean_value_jac_am = result_frame["ama_jaccard_sim"].mean()

print("max_value_jaro:", max_value_jaro_am,"min_value_jaro:", min_value_jaro_am)
print("max_value_lev:", max_value_lev_am, "min_value_lev:",min_value_lev_am)
print("max_value_jac:", max_value_jac_am,"min_value_jac:", min_value_jac_am)
```

```python
ama_res = pd.DataFrame(data=[], columns=['online_shop', 'amazon', 'ama_jarowinkler_sim', 'ama_levenshtein_sim', 'ama_jaccard'])

def amazon_result(df):
    for ind in df.index: 
        if (df['ama_jarowinkler_sim'][ind] == max_value_jaro_am):
            ama_res.loc[ind] = [df['online_shop'][ind], df['amazon'][ind], df['ama_jarowinkler_sim'][ind], df['ama_levenshtein_sim'][ind], df['ama_jaccard_sim'][ind]]
```

```python
amazon_result(result_frame)
```
